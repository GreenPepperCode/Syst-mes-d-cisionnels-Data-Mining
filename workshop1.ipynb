```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

# Importation du dataset
store_data = pd.read_csv('C:\\Users\\yzi\\Desktop\\Travaux\\Systemes-decisionnels-Data-Mining\\data\\store_data.csv', header=None)

# Prétraitement des données
records = []
for i in range(0, 7501):
    records.append([str(store_data.values[i, j]) for j in range(0, 20) if str(store_data.values[i, j]) != 'nan'])

# Application de l'algorithme Apriori
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)

# Affichage des résultats
for item in association_results:
    # Première partie de chaque règle
    pair = item[0]
    items = [x for x in pair]
    print("Règle: " + items[0] + " -> " + items[1])

    # Support
    print("Support: " + str(item[1]))

    # Confiance
    print("Confiance: " + str(item[2][0][2]))

    # Lift
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
